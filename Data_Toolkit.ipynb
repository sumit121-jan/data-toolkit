2. How does broadcasting work in NumPy ?
- Broadcasting is a feature in NumPy that allows operations on arrays of different shapes without explicitly reshaping or replicating data. It enables element-wise operations efficiently, avoiding unnecessary memory usage.
- Rules of Broadcasting >>
 - Matching Dimensions - If the arrays have the same shape, operations are performed element-wise.
 - Expanding Singleton Dimensions - If an array has a dimension of size 1, NumPy stretches it to match the other array.
 - Padding Smaller Arrays - If one array has fewer dimensions, NumPy adds leading dimensions with size 1 until both arrays have the same number of dimensions.

6. What are the differences between NumPy arrays and Python lists ?
- Difference between NumPy arrays and Python lists >>
  - Both NumPy arrays and Python lists store collections of elements, but NumPy arrays are specifically optimized for numerical computations. Here’s how they differ >>
  - Performance - NumPy arrays are faster than Python lists because they use contiguous memory blocks and are implemented in C and Fortran.Python lists store objects individually, making them slower for large-scale operations.
  - Memory Efficiency - NumPy arrays use less memory since they store elements as fixed data types (e.g., int32, float64).Python lists store references to objects, requiring more memory.
  - Functionality & Operations - NumPy supports vectorized operations, meaning mathematical operations apply to the entire array at once.Python lists require loops, making operations slower.
  - Data Type Flexibility - NumPy arrays require elements of the same data type (e.g., all integers or all floats).Python lists can store mixed data types (e.g., integers, strings, floats in the same list).
  - Multi-Dimensional Support - NumPy supports multi-dimensional arrays (e.g., matrices, tensors).Python lists are only one-dimensional by default and require nested lists for multi-dimensional structures.

8. What does the term “vectorized operation” mean in NumPy ?
- A vectorized operation in NumPy refers to the process of performing mathematical operations on entire arrays (vectors, matrices, or tensors) at once, without using explicit loops. Instead of iterating over each element individually, as is typically done with Python lists, NumPy applies operations simultaneously to all elements in the array using highly optimized, low-level C and Fortran implementations.
- This approach significantly improves performance by leveraging parallel computation and efficient memory handling, making vectorized operations much faster than traditional loops in Python.

17. Explain the difference between apply() and map() in Pandas .
- Differences between apply() and map() >>
- apply() Function - The apply() function is used to apply a function along an axis of a DataFrame or Series. It can be used to apply functions to rows, columns, or elements of the DataFrame/Series
 - Data Type Compatibility - apply() works with both DataFrames and Series.
 - Flexibility -  It allows you to apply functions to either rows or columns (using the axis argument).
 - Return Type - The return type can vary; it could be a Series, DataFrame, or scalar value, depending on the function applied.

- map() Function - The map() function is used for element-wise transformations on a Series. It allows you to map each value in a Series to another value using a function, dictionary, or Series.
  - Data Type Compatibility - map() works only with Series.
  - Flexibility - It is mainly used for element-wise transformations and cannot operate across rows or columns in a DataFrame.
  - Return Type -  The return is always a Series, and it performs an element-wise operation.

- apply() is more versatile and works on both DataFrames and Series, allowing for row-wise or column-wise operations. It's often used for more complex operations or when a custom function needs to be applied.
- map() is used for element-wise transformations on a Series, typically for simpler tasks like replacing values or applying a function to each element.

18. What are some advanced features of NumPy ?
- Here are some of the advanced features of NumPy >>

  - Linear Algebra Functions - NumPy provides comprehensive support for linear algebra operations, such as matrix multiplication, eigenvalue decomposition, and solving linear systems.
  - Universal Functions (ufuncs) - NumPy provides universal functions (ufuncs), which are functions that operate on arrays element-wise. These are implemented in C, making them much faster than writing loops in Python.
  -  Random Module - The np.random module in NumPy provides a variety of functions to generate random numbers and samples from different distributions
  - Memory Layouts and Data Types - NumPy allows users to control the memory layout of arrays for performance optimization. You can use structured arrays to handle data with mixed types (e.g., integers and floats) and record arrays for efficient storage of heterogeneous data.
  - Fourier Transform (FFT) - NumPy provides tools for performing Fourier transforms to analyze signals in the frequency domain. This is useful in areas like signal processing, audio analysis, and image processing.
  - Data Wrangling and Aggregation - NumPy supports advanced aggregation and data wrangling capabilities, such as grouping, splitting, and applying functions. This is useful when working with large datasets.

21. Why is NumPy’s array slicing faster than Python’s list slicing ?
- NumPy's array slicing is faster than Python's list slicing because of several key differences in how NumPy arrays and Python lists are implemented and optimized:
  - Contiguous Memory Allocation - NumPy arrays store data in contiguous blocks of memory, allowing faster access to slices.
Python lists store references to objects, often in non-contiguous memory locations, which adds overhead.
  - Memory Views vs. Copies - NumPy slicing creates a view of the original data, referencing the same memory without duplicating data.
Python list slicing creates a new copy of the sliced data, which is slower due to the copying process.
  - Vectorized Operations - NumPy supports vectorized operations, applying optimizations in C to operate on entire arrays at once.
Python lists do not support vectorized operations and must iterate through elements one by one.
  - Low-Level Optimizations - NumPy is built on optimized C libraries and hardware-level optimizations for fast computation.
Python lists are managed by the Python interpreter, which adds overhead to slicing.
  - Memory Efficiency - NumPy arrays use fixed-size data types, allowing for better memory usage and faster access.
Python lists store references to Python objects, which use more memory and reduce efficiency.

```python
#4 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 75000, 80000, 90000],
    'Experience': [2, 5, 7, 10, 12],
    'Work Hours': [40, 42, 45, 50, 55]
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(8, 6))  # Set figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add a title
plt.title("Correlation Matrix Heatmap")

# Display the heatmap
plt.show()
```

```python
#11 Perform matrix multiplication using NumPy.
import numpy as np

# Define two matrices
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix multiplication using @ operator
result1 = A @ B

# Perform matrix multiplication using np.matmul()
result2 = np.matmul(A, B)

# Print the results
print("Result using @ operator:\n", result1)
print("\nResult using np.matmul():\n", result2)
```
